### Code pour la classification avec SVM

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, ConfusionMatrixDisplay
```

```python
# 1. Chargement des données (ex: jeu de données Iris)
iris = datasets.load_iris()
X, y = iris.data[:, :2], iris.target  # On utilise seulement 2 features pour la visualisation
```

```python
X
```

```python
y
```

```python
# 2. Séparation en ensemble d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# 3. Normalisation des données
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# 4. Création et entraînement du modèle SVM
svm_model = SVC(kernel='rbf', C=1.0, gamma='scale')  # Kernel radial basis function (RBF)
svm_model.fit(X_train, y_train)
```

```python
# 5. Prédiction et évaluation
y_pred = svm_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
```

```python
print(f"Précision du modèle : {accuracy:.2f}")
print("Rapport de classification :\n", classification_report(y_test, y_pred))
```

```python
# 6. Matrice de confusion
ConfusionMatrixDisplay.from_estimator(svm_model, X_test, y_test)
plt.show()
```

 Explication du code :
On utilise le jeu de données Iris.

On sélectionne 2 caractéristiques pour faciliter la visualisation.

On divise les données en train/test.

On applique une normalisation avec StandardScaler().

On entraîne un modèle SVM avec un kernel RBF.

On évalue le modèle avec la précision et une matrice de confusion.
