###  Exemple d'AdaBoost pour classification (avec scikit-learn)

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

# 1. Générer un dataset de classification

```python
X, y = make_classification(n_samples=1000, n_features=20,
                           n_informative=15, n_redundant=5,
                           random_state=42)
```

```python
X
```

```python
y
```

# 2. Diviser en ensemble d'entraînement et de test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3,
                                                    random_state=42)
```

# 3. Créer un classifieur de base (stump: arbre de décision de profondeur 1)

```python
base_estimator = DecisionTreeClassifier(max_depth=1)
```

# 4. Créer et entraîner l'AdaBoost

```python
ada = AdaBoostClassifier(base_estimator,
                         n_estimators=50,
                         learning_rate=1.0,
                         random_state=42)
```

```python
ada.fit(X_train, y_train)
```

# 5. Prédictions et évaluation

```python
y_pred = ada.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
